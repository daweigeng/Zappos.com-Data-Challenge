```python
def index_plot(df,title,subplot=True,xlabel="Date",site={"Acme":'g',"Botly":'k',"Pinnacle":'y',"Sortly":'m'},other={"Tabular":'c',"Widgetry":'r'}):
    ax=0
    if subplot:
        plt.subplot(2, 1, 1)
    else:
        site.update(other)
    for i,value in site.items():
        df_site=df[df['site']==i]
        df_site.index=list(df_site.groupby('day').apply(lambda x: datetime.datetime.strptime(x.iloc[0]['day'].split(" ")[0], '%m/%d/%Y').strftime('%Y-%m-%d')).reset_index()[0])
        #df.sort_index(inplace=True)
        dates = pd.date_range('2013-01-01', '2013-12-31')
        df1 = pd.DataFrame(index=dates)
        df_all_date = df1.join(df_site)
        if type(ax)!="matplotlib.axes._subplots.AxesSubplot":
            ax = df_all_date[title].plot(title=title,color=value,fontsize=12,label=i)
        else:
            df_all_date[title].plot(ax=ax,color=value,label=i)
    ax.legend(bbox_to_anchor=(1.31, 0.6))

    
    if subplot:
        plt.subplot(2,1,2)
        ax1=0
        for key,value in other.items():
            df_site=df[df['site']==key]
            df_site.index=list(df_site.groupby('day').apply(lambda x: datetime.datetime.strptime(x.iloc[0]['day'].split(" ")[0], '%m/%d/%Y').strftime('%Y-%m-%d')).reset_index()[0])
            #df.sort_index(inplace=True)
            dates = pd.date_range('2013-01-01', '2013-12-31')
            df1 = pd.DataFrame(index=dates)
            df_all_date = df1.join(df_site)
            if type(ax1)!="matplotlib.axes._subplots.AxesSubplot":
                ax1 = df_all_date[title].plot(color=value,fontsize=12,label=key)
            else:
                df_all_date[title].plot(ax=ax1,color=value,label=key)
        ax1.legend(bbox_to_anchor=(1.31, 0.6))
        ax1.set_xlabel(xlabel)
    else:
        ax.set_xlabel(xlabel)
        ax.set_ylabel(title)
    plt.tight_layout()
    plt.show()
```

```python
def plotly_plot_data(df,target,site=["Acme","Pinnacle","Sortly"]):
    dates = pd.date_range('2013-01-01', '2013-12-31')
    df1 = pd.DataFrame(index=dates)
    for i in site:
        df_site=df[df['site']==i]
        df_site.index=list(df_site.groupby('day').apply(lambda x: datetime.datetime.strptime(x.iloc[0]['day'].split(" ")[0], '%m/%d/%Y').strftime('%Y-%m-%d')).reset_index()[0])
        #df.sort_index(inplace=True) 
        df2=df_site[target]
        df2.name=i
        df1 = df1.join(df2)    
    return df1

     
```

```python
def Time_Series_Plotly(df1,target,site=["Acme","Pinnacle","Sortly"]):
    data=[]
    for i in site:
        data.append(go.Scatter(
            x=list(df1.index),
            y=df1[i],
            name=i
        )) 
    layout = go.Layout(
        title='Time Series Plot for '+target,
        xaxis=dict(
            title='Date',
            titlefont=dict(
                family='Courier New, monospace',
                size=18,
                color='#7f7f7f'
            )
        ),
        yaxis=dict(
            title=target,
            titlefont=dict(
                family='Courier New, monospace',
                size=18,
                color='#7f7f7f'
            )
        )
    )
    fig = go.Figure(data=data, layout=layout)
    plot_url = py.iplot(fig)
    return plot_url
```

```python
def plotly_plot_data_platform(df,target,site=['Android','Linux','Macintosh','Other','Windows','iPad','iPhone','BlackBerry','ChromeOS','MacOSX','WindowsPhone','iOS','Unknown','SymbianOS']):
    dates = pd.date_range('2013-01-01', '2013-12-31')
    df1 = pd.DataFrame(index=dates)
    for i in site:
        df_site=df[df['platform']==i]
        df_site.index=list(df_site.groupby('day').apply(lambda x: datetime.datetime.strptime(x.iloc[0]['day'].split(" ")[0], '%m/%d/%Y').strftime('%Y-%m-%d')).reset_index()[0])
        #df.sort_index(inplace=True) 
        df2=df_site[target]
        df2.name=i
        df1 = df1.join(df2)    
    return df1
```

```python
def plotly_plot_data_stacked_bar(df,target,site=['Android','Linux','Macintosh','Other','Windows','iPad','iPhone','BlackBerry','ChromeOS','MacOSX','WindowsPhone','iOS','Unknown','SymbianOS']):
    dates = range(7)
    df1 = pd.DataFrame(index=dates)
    for i in site:
        df_site=df[df['platform']==i]
        dd=df_site.groupby('day').apply(lambda x: datetime.datetime.strptime(x.iloc[0]['day'].split(" ")[0], '%m/%d/%Y').weekday()).reset_index()
        dd=dd.rename(columns={'day':'day',0:'weekday'})
        df_site=df_site.merge(dd,on='day')
        df_site=df_site.groupby('weekday')[target].sum().reset_index()
        df_site.index=df_site.weekday
        df_site=df_site[target]
        df_site.name=i
        df1 = df1.join(df_site)    
    return df1
def plotly_plot_data_stacked_month(df,target,site=['Android','Linux','Macintosh','Other','Windows','iPad','iPhone','BlackBerry','ChromeOS','MacOSX','WindowsPhone','iOS','Unknown','SymbianOS']):
    dates = range(1,13)
    df1 = pd.DataFrame(index=dates)
    for i in site:
        df_site=df[df['platform']==i]
        dd=df_site.groupby('day').apply(lambda x: int(datetime.datetime.strptime(x.iloc[0]['day'].split(" ")[0], '%m/%d/%Y').strftime('%m'))).reset_index()
        dd=dd.rename(columns={'day':'day',0:'month'})
        df_site=df_site.merge(dd,on='day')
        df_site=df_site.groupby('month')[target].sum().reset_index()
        df_site.index=df_site.month
        df_site=df_site[target]
        df_site.name=i
        df1 = df1.join(df_site)    
    return df1
```

```python
def df_month_or_week(data,col,month=True):
    if month:
        df_month=plotly_plot_data_stacked_month(data8,col)
        df_month['total']=df_month.sum(axis=1)
        df_month.index=['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
    else:
        df_month=plotly_plot_data_stacked_bar(data8,col)
        df_month['total']=df_month.sum(axis=1)
        df_month.index=['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
    return df_month
```
